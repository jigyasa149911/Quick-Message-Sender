# Quick Message Sender

```python
import requests
import json
```

The function below will take the mobile number and message from the GUI and will use Fast2SMS API server to send messages.
The parameters will be given in form of a dictionary, where authorization will be a unique key generated for your account, message will be message passed to the send_message function, language will be engluish, route will be 'q' to send quick messages and number parameter will contain the number where message needs to be sent.

```python
def send_message(number, message):
    # URL to access Fast2SMS server by GET request.
    url = 'https://www.fast2sms.com/dev/bulkV2'
    
    parameters = {
        'authorization':'Insert your authorization id of the API',
        
        'message':message,
        'language':'english',
        
        # Use 'q' for quick message with default 
        'route':'q',
        
        # The number where we have to send the message, we can use various comma separated numbers to send the message,
        # here we are using only one number, that is already passed in send_message() function.
        'numbers': number
        
        }
    
    response = requests.get(url, params = parameters)

    # We are changing the response into json and saving it to a dictionary.
    dict = response.json() 
    print(dict)
    return dict['return'] # It will return true if message is sent.
```

Let's import showinfo() and showerror() functions to show success and failure messages.

```python
from tkinter.messagebox import showinfo, showerror
```

The function below will define the event that will take place when we are going to press the send button in our GUI.

```python
def button_click():
    num = textNumber.get()
    msg = textMessage.get("1.0",END) # Saying that we want complete message from index 1 to END.
    # Calling send_message function
    sent = send_message(num, msg)
    
    #Checking if message is sent or not.
    if sent:
        # This method will be used for showing information.
        showinfo("Sent Successfully","Successfully Sent")
    else:
        # This method will be used for showing error, if occurred during the process.
        showerror("Error","Something went wrong")
```

# Creating GUI of Quick Message Sender

```python
from tkinter import *
```

```python
# Creating the main window of GUI
root = Tk()

# Title of the GUI
root.title("QUICK MESSAGE SENDER")

# Setting the dimension of our window.
root.geometry("400x550")

# Setting the font of GUI. ("font_family", font_size, "font_style")
font= ("Helvetica",22,"italic")

# Text Field
textNumber = Entry(root, font= font)
#padx = padding of x axis, pady = padding of y axis
textNumber.pack(pady = 20, padx=20)

# Making text box for entering message.
textMessage = Text(root)
textMessage.pack(pady = 20, padx = 20)

# Making the submit button.
sendButton = Button(root, text = "SEND MESSAGE", command = button_click)
sendButton.pack()

root.mainloop()
```
